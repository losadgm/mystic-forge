```python
import pandas as pd

df = pd.read_csv("../data/raw/cards_seed.csv")
```

```python
print(df.shape)
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
print(df.head())
```

```python
cols = ["name", "type", "mana_cost", "power", "toughness", "rarity", "colors"]
df = df[cols]
```

```python
df = df[pd.to_numeric(df["power"], errors="coerce").notna()]
df = df[pd.to_numeric(df["toughness"], errors="coerce").notna()]
df["power"]     = df["power"].astype(int)
df["toughness"] = df["toughness"].astype(int)
print(df.dtypes)
```

```python
df["colors"] = df["colors"].fillna("Colorless")
df["has_mana_cost"] = df["mana_cost"].notna()
df["mana_cost"] = df["mana_cost"].fillna("none")

#print(df[df["colors"] == "Colorless"].head())
print(df[df["mana_cost"] == "none"].head())
```
